### 1.1. Raw Train Dataset Overview
Shape, columns overview, and some essential statistics, and a peak into a few entries:

```python
def header(width: int, title: str) -> str:
    """
    Just a function to generate a pretty header for our output data
    :param width: The width (in characters) of the header
    :param title: The content (title) of the header
    :return: The generated header, as a string
    """
    calculated_width = width if len(title) + 4 <= width else len(title) + 4  # Auto padding
    border_top = '╔' + '═'*(calculated_width - 2) + '╗'
    border_bottom = '╚' + '═'*(calculated_width - 2) + '╝'
    border_vertical = '║'
    return f"{border_top}\n{border_vertical}{title : ^{calculated_width - 2}}{border_vertical}\n{border_bottom}"

print(f"{header(9, 'SHAPE')}\n{df_train.shape}")
print(f"\n{header(19, 'NULL VALUES COUNT')}\n{df_train.isna().sum()}")
print(f"\n{header(39, 'COLUMNS OVERVIEW')}")
print(df_train.info())
```

### 1.3 Observation

The target column we are trying to predict is `Sepsis` (Positive/Negative) which records if a patient **had developed** sepsis during their ICU stay.

The 8 features we will be using to predict the target are:

| Feature  | Data Type  | Description                                    |
|:---------|:-----------|:-----------------------------------------------|
| PRG      | int        | Plasma glucose                                 |
| PL       | int        | Blood Work Result-1 (mu U/ml)                  |
| PR       | int        | Blood Pressure (mm Hg)                         |
| SK       | int        | Blood Work Result-2 (mm)                       |
| TS       | int        | Blood Work Result-3 (mu U/ml)                  |
| M11      | float      | Body mass index (weight in kg/(height in m)^2  |
| BD2      | float      | Blood Work Result-4 (mu U/ml)                  |
| Age      | int        | Patient's age (years)                          |

Unused columns: `ID` and `Insurance`. These have no connection to the patient physical well-being.

```python
df_train = df_train.drop(columns=['ID', 'Insurance'])
df_test = df_test.drop(columns=['ID', 'Insurance'])
```

Missing attribute values: **Not found in both datasets**. Thus, we don't need to handle missing data.

**Conclusion:** We will be predicting `Sepsis` for new unseen data using the above 8 features. To do so, our prediction will be `0 < x < 1` where `x` is the probability of a patient developing sepsis during their ICU stay.

**Note:** To train our model, we need to convert the type of the `Sepsis` column in the train dataset to an integer value with `0` corresponds to `Negative` and `1` corresponds to `Positive`.

```python
df_train_before = df_train.copy()
df_train["Sepsis"] = df_train["Sepsis"].map({"Negative": 0, "Positive": 1})
```

## 2. Data Distribution

### 2.1. Overview
We will observe the data distribution by plotting a histogram for each column of our datasets. The test dataset is overlaid on top of the train dataset, except for the Sepsis column because it is only available in the Train dataset. KDE lines for both train and test data are also included to better visualize their distribution.

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15), constrained_layout=True)
axes = axes.ravel()

for i, col in enumerate(df_train.columns):
    if col != "Sepsis":
        axes[i].hist(
            df_train[col],
            bins=min(df_train.shape[0] // 10, 18),
            color="royalblue",
            density=True,
            alpha=0.5,
            label="Train Data"
        )
        sns.kdeplot(df_train[col], ax=axes[i], color="blue", alpha=0.8, label="Train Data KDE")

        axes[i].hist(
            df_test[col],
            bins=min(df_test.shape[0] // 10, 18),
            color="crimson",
            density=True,
            alpha=0.5,
            label="Test Data"
        )
        sns.kdeplot(df_test[col], ax=axes[i], color="red", alpha=0.8, label="Test Data KDE")

        axes[i].set_ylabel("Probability density", fontsize=10)
        axes[i].set_xlabel("Value", fontsize=10)
        axes[i].set_xlim(left=min(df_train[col].min(), df_test[col].min()), right=max(df_train[col].max(), df_test[col].max()))
    else:
        axes[i].hist(
            df_train[col],
            bins=2,
            color="sandybrown",
            alpha=0.8,
            label="Train Data Target"
        )
        axes[i].set_ylabel("Count", fontsize=10)
        axes[i].set_xticks([0.25, 0.75], labels=["Negative", "Positive"])
    axes[i].tick_params(axis="both", which="major", labelsize=10)
    axes[i].set_title(col, fontsize=14, fontweight="bold")
    axes[i].legend()

plt.show()

fig.savefig("../images/EDA_Distribution_Overview.png", dpi=300)
```

### 2.1.1. Distribution Of PRG Feature
Plasma glucose.

```python
def show_distribution(col: str, train: pd.DataFrame, test: pd.DataFrame, save_path: str=None, save_dpi: int=300) -> None:
    """
    Displays comparison histograms and accompanying box plots for col of both train and test DataFrame. Then save that plot into the specified path, if it exists.
    :param col: Name of the column to be analyzed.
    :param train: train DataFrame.
    :param test: test DataFrame.
    :param save_path: The location for the plot produced by this function to be saved at. Defaults to None (no saving).
    :param save_dpi: The quality of the saved plot. Only takes effect if save_path is not None.
    :return: None
    """
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7, 5), constrained_layout=True, sharex="col", gridspec_kw={"height_ratios": [3, 1]})

    # Graph 1: Histogram
    axes[0, 0].set_title(f"{col} Distribution - Train Data")
    axes[0, 0].set_ylabel("Count")
    axes[0, 0].hist(
        train[col],
        bins=min(train.shape[0] // 10, 24),
        color="royalblue",
        alpha=0.9,
    )

    axes[0, 1].set_title(f"{col} Distribution - Test Data")
    axes[0, 1].set_ylabel("Count")
    axes[0, 1].hist(
        test[col],
        bins=min(test.shape[0] // 10, 24),
        color="crimson",
        alpha=0.8,
    )

    # Graph 2: Box plot
    axes[1, 0].boxplot(train[col], vert=False)
    axes[1, 0].set_yticks([])
    axes[1, 1].boxplot(test[col], vert=False)
    axes[1, 1].set_yticks([])
    plt.show()

    # Print out numerical distribution statistics
    df_prg_summary = pd.concat([train[[col]].describe(), test[[col]].describe()], axis=1)
    df_prg_summary.columns = [f"{col}_TRAIN", f"{col}_TEST"]
    print(df_prg_summary)

    if save_path is not None:
        fig.savefig(f"../images/EDA_Distribution_{col}.png", dpi=300)

show_distribution("PRG", df_train, df_test, "../images/EDA_Distribution_PRG.png")
```

## 3. Feature Correlation

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 7))

axes[0].set_title("Correlation Matrix for Train Data", fontsize=16, fontweight="bold")
mask = np.zeros_like(df_train.corr())
sns.heatmap(df_train.corr(), vmin=-1, vmax=1, cmap='RdBu_r', annot=True, ax=axes[0], annot_kws={"fontsize":12}, xticklabels=True)
axes[0].tick_params(axis='both', which='major', labelsize=12)

axes[1].set_title("Correlation Matrix for Test Data", fontsize=16, fontweight="bold")
sns.heatmap(df_test.corr(), vmin=-1, vmax=1, cmap='RdBu_r', annot=True, ax=axes[1], annot_kws={"fontsize":12}, xticklabels=True)
axes[1].tick_params(axis='both', which='major', labelsize=12)

fig.savefig("../images/EDA_Correlation.png", dpi=300)

plt.show()
```

From the above correlation matrices, we can conclude that both the train data and test data has very similar correlation between features. This further re-enforce our confidence that our model trained on the train dataset will perform well on the test dataset.

In more details:

- There is a moderate (0.53) correlation of Age-PRG. This is in agreement with medical researches done across a wide spectrum of races and sexes regarding the positive correlation between age and plasma glucose level due to the declining ability of the aging body to regulate blood glucose level [\[1\]](https://www.researchgate.net/publication/6753652_Effects_of_Age_on_Plasma_Glucose_Levels_in_Non-diabetic_Hong_Kong_Chinese) [\[2\]](https://www.ahajournals.org/doi/10.1161/CIRCRESAHA.118.312806) [\[3\]](https://bmcpublichealth.biomedcentral.com/articles/10.1186/s12889-021-10367-x) [\[4\]](https://www.researchgate.net/publication/329474872_The_effects_of_age_and_body_mass_index_on_blood_glucose_blood_cholesterol_and_blood_pressure_in_adult_women).
- As such, correlation of PRG-Sepsis and Age-Sepsis are roughly the same (~0.21), which is beneficial for training our model. We would expect these two features to have moderately significant weight on our model, and we might be able to drop one of those two features without affecting our model's performance.
- We also see a cluster of correlation between TS, PL, PR, M11, and BD2, but none of them are as significant as the correlation of Age-PRG. We can ignore this for now, unless we can theoretically or empirically prove that these correlation are significant enough to warrant manual feature selection.
- Aside from Age-Sepsis and PRG-Sepsis, we also see significant correlations (>0.1) of PL-Sepsis, M11-Sepsis, BD2-Sepsis, and TS-Sepsis (listed in order of level of correlation). We can also expect these feature to have significant weight in our model.
